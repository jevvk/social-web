Our dataset is preprocessed in data/datasets.db

```python
%matplotlib notebook

import sqlite3
import matplotlib.pyplot as plt
import numpy as np

# get the db
conn = sqlite3.connect('data/datasets.db')
c = conn.cursor()
```

Question 1a: Is there a difference in profile picture usage?

```python
age1 = []   #age range 1 20-40
age2 = []   #age range 2 40-60
age3 = []   #age range 3 60-80

#Running queries
for gender in ('male', 'female', ''):
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=20 and ageEstimate<40 and hasPicture<>"" and genderEstimate = ?', [gender])
    res1 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=40 and ageEstimate<60 and hasPicture<>"" and genderEstimate = ?', [gender])
    res2 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=60 and ageEstimate<=86 and hasPicture<>"" and genderEstimate = ?', [gender])
    res3 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=20 and ageEstimate<40 and genderEstimate = ? ', [gender])
    total1 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=40 and ageEstimate<60 and genderEstimate = ? ', [gender])
    total2 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=60 and ageEstimate<=86 and genderEstimate = ? ', [gender])
    total3 = int(c.fetchall()[0][0])
  
    age1.append(res1 / total1 * 100)
    age2.append(res2 / total2 * 100)
    age3.append(res3 / total3 * 100)
    
barWidth = 0.25 # the width of the bars
positions = np.arange(len(age1))

# Make the plot
plt.bar(positions, age1, color='#008000', width=barWidth, edgecolor='white', label='20-40')
plt.bar(positions + barWidth, age2, color='#e60000', width=barWidth, edgecolor='white', label='40-60')
plt.bar(positions + barWidth*2, age3, color='#0000e6', width=barWidth, edgecolor='white', label='60-80')

plt.xlabel('Genders', fontweight='bold')
plt.xticks(positions + barWidth, ['Male', 'Female', 'Unknown'])
plt.title('Picture usage')

plt.legend()
plt.show()
```

Question 2: How do companies compare in popularity between LinkedIn and Facebook?

```python
# hp repeats
c.execute("select company_name, followers_count from linkedin_dataset where company_name<>'Hewlett-Packard' order by followers_count desc limit 10")
res1 = c.fetchall()
# mcdonalds repeats
c.execute("select username, likes from facebook_dataset where username<>'McDonaldsUS' order by likes desc limit 10")
res2 = c.fetchall()

fig = plt.figure(dpi=110)

ax = fig.add_subplot(1,2,1)
table = ax.table(cellText=res1, colLabels=['Company', 'Number of Followers'], loc='center')
table.set_fontsize(25)
table.scale(1,2)
ax.axis('off')

ax = fig.add_subplot(1,2,2)
table = ax.table(cellText=res2, colLabels=['Company', 'Number of Likes'], loc='center')
table.set_fontsize(25)
table.scale(1,2)
ax.axis('off')

plt.show()
```

Question 2a: Is there a correlation between industry and popularity of companies per platform?

```python
c.execute('select industry, avg(likes), avg(followers_count) from linkedin_dataset join facebook_dataset' + \
          ' on company_name = username where industry in ("Information Technology and Services", "Internet", "Computer Software", "Retail", "Financial Services") group by industry')
res = c.fetchall()

industries = list(map(lambda x: x[0], res))
industries[industries.index('Information Technology and Services')] = 'IT & Services' # shorten name for graph
likes = list(map(lambda x: x[1], res))
followers = list(map(lambda x: x[2], res))

barWidth = 0.25 # the width of the bars
# set height of bar
bars1 = followers
bars2 = likes

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Make the plot  
plt.bar(r1, bars1, color='#283e4a', width=barWidth, edgecolor='white', label='followers')
plt.bar(r2, bars2, color='#3d6ad6', width=barWidth, edgecolor='white', label='likes')

# Add xticks on the middle of the group bars   
plt.xlabel('Industries', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], industries)
plt.title('Followers count and likes for the most popular industries')

# Create legend & Show graphic
plt.legend()
plt.show()
```

Question 2b: Is there a correlation between the size and popularity of  companies per platform?

```python
followers_count = []
likes = []
#Running queries
#temp variable stores the query's outcome
#importing query's values (via temp variable) to the followers_count variable
c.execute("select avg(followers_count) from linkedin_dataset, facebook_dataset where company_name= username and employees_on_platform>=0 and employees_on_platform<50")
temp = c.fetchall()
followers_count.append(int(temp[0][0]))
c.execute("select avg(followers_count) from linkedin_dataset, facebook_dataset where company_name= username and employees_on_platform>=50 and employees_on_platform<250")
temp = c.fetchall()
followers_count.append(int(temp[0][0]))
c.execute("select avg(followers_count) from linkedin_dataset, facebook_dataset where company_name= username and employees_on_platform>=250 and employees_on_platform<500")
temp = c.fetchall()
followers_count.append(int(temp[0][0]))
c.execute("select avg(followers_count) from linkedin_dataset, facebook_dataset where company_name= username and employees_on_platform>=500 and employees_on_platform<=1000")
temp = c.fetchall()
followers_count.append(int(temp[0][0]))
#importing query's values (via temp variable) to the likes variable
c.execute("select avg(likes) from linkedin_dataset, facebook_dataset where company_name= username and employees_on_platform>=0 and employees_on_platform<50")
temp = c.fetchall()
likes.append(int(temp[0][0]))
c.execute("select avg(likes) from linkedin_dataset, facebook_dataset where company_name= username and employees_on_platform>=50 and employees_on_platform<250")
temp = c.fetchall()
likes.append(int(temp[0][0]))
c.execute("select avg(likes) from linkedin_dataset, facebook_dataset where company_name= username and employees_on_platform>=250 and employees_on_platform<500")
temp = c.fetchall()
likes.append(int(temp[0][0]))
c.execute("select avg(likes) from linkedin_dataset, facebook_dataset where company_name= username and employees_on_platform>=500 and employees_on_platform<=1000")
temp = c.fetchall()
likes.append(int(temp[0][0]))

barWidth = 0.25 # the width of the bars
# set height of bar
bars1 = followers_count
bars2 = likes

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Make the plot   
plt.bar(r1, bars1, color='#283e4a', width=barWidth, edgecolor='white', label='Followers Count')
plt.bar(r2, bars2, color='#3d6ad6', width=barWidth, edgecolor='white', label='Likes')

# Add xticks on the middle of the group bars   
plt.xlabel('Company Size', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['0-50', '50-250', '250-500', '500-1000'])
plt.title('Followers and Likes Average Seperated By Company Size')

# Create legend & Show graphic
plt.legend()
plt.show()

```
