```python
%matplotlib notebook

import sqlite3
import matplotlib.pyplot as plt
import numpy as np

# get the db
conn = sqlite3.connect('data/datasets.db')
c = conn.cursor()
```

Question 1: 

```python
age1 = []   #age range 1 20-40
age2 = []   #age range 2 40-60
age3 = []   #age range 3 60-80

#Running queries
for gender in ('male', 'female', ''):
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=20 and ageEstimate<40 and hasPicture<>"" and genderEstimate = ?', [gender])
    res1 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=40 and ageEstimate<60 and hasPicture<>"" and genderEstimate = ?', [gender])
    res2 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=60 and ageEstimate<=86 and hasPicture<>"" and genderEstimate = ?', [gender])
    res3 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=20 and ageEstimate<40 and genderEstimate = ? ', [gender])
    total1 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=40 and ageEstimate<60 and genderEstimate = ? ', [gender])
    total2 = int(c.fetchall()[0][0])
    c.execute('select count(*) from kaggle_dataset where ageEstimate>=60 and ageEstimate<=86 and genderEstimate = ? ', [gender])
    total3 = int(c.fetchall()[0][0])
  
    age1.append(res1 / total1 * 100)
    age2.append(res2 / total2 * 100)
    age3.append(res3 / total3 * 100)
    
barWidth = 0.25 # the width of the bars
positions = np.arange(len(age1))

# Make the plot
plt.bar(positions, age1, color='#008000', width=barWidth, edgecolor='white', label='20-40')
plt.bar(positions + barWidth, age2, color='#e60000', width=barWidth, edgecolor='white', label='40-60')
plt.bar(positions + barWidth*2, age3, color='#0000e6', width=barWidth, edgecolor='white', label='60-80')

plt.xlabel('Genders', fontweight='bold')
plt.xticks(positions + barWidth, ['Male', 'Female', 'Unknown'])
plt.title('Picture usage')

plt.legend()
plt.show()
```
